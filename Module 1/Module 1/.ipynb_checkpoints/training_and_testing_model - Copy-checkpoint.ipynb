```python
disease_symptom_counts = df.groupby('disease').sum()
disease_symptom_counts
```

```python
# Normalize the counts to calculate priority
disease_symptom_priority = disease_symptom_counts / disease_symptom_counts.sum()
disease_symptom_priority
```

```python
# Create columns for each symptom's priority for each disease
for symptom in df.columns[:-1]:
    df[symptom + '_Priority'] = df['disease'].map(disease_symptom_priority[symptom])
```
